```python
# Imported Libraries for optimization of code
import sys
import heapq

# defining a class max_heap for the heap data structure
# Using OOP Methodologies
class max_heap: # defined class
    def __init__(self, sizelimit): # instanciate self
        self.sizelimit = sizelimit
        self.cur_size = 0
        self.Heap = [0]*(self.sizelimit + 1)
        self.Heap[0] = sys.maxsize
        self.root = 1
# assigned all vaiables with self to make use in functions of this class

    # helper function to swap the two given nodes of the heap
    # this function will be needed for max-heapify and insertion 
    # in order to swap nodes which are not in order (not satisfy max-heap property)
    def swapnodes(self, node1, node2):
        self.Heap[node1], self.Heap[node2] = self.Heap[node2], self.Heap[node1] # Swapping two  nodes
    
    # THE MAX_HEAPIFY FUNCTION taking i in argument
    def max_heapify(self, i):
 
        # If the node is a not a leaf node and is lesser than any of its child
        if not (i >= (self.cur_size//2) and i <= self.cur_size):
            if (self.Heap[i] < self.Heap[2 * i]  or  self.Heap[i] < self.Heap[(2 * i) + 1]): 
                if self.Heap[2 * i] > self.Heap[(2 * i) + 1]:
     # Swap the node with the left child and call the max_heapify function on it
                    self.swapnodes(i, 2 * i)
                    self.max_heapify(2 * i)
     # Used swapnodes method which we declared above to swap nodes
                else:
                # Swap the node with right child and then call the max_heapify function on it
                    self.swapnodes(i, (2 * i) + 1)
                    self.max_heapify((2 * i) + 1)
 


    # THE Insert_MaxHeap Function to insert new node
    def Insert_MaxHeap(self, element): # takes 1 element argument
        if self.cur_size >= self.sizelimit : # check ifcurrent size is greater than overallsize
            return # means this is first node
        self.cur_size+= 1 # addition in size
        self.Heap[self.cur_size] = element  # assign size to element
        current = self.cur_size
        while self.Heap[current] > self.Heap[current//2]: # while cureent node value is greter than its parent
            self.swapnodes(current, current//2) # swap the values place on those indexes
            current = current//2 # assign that value to current variable
 
 
    # THE Build_Heap_Max FUNCTION to Build heap when max node removed
    def Build_Heap_Max(self):
        last = self.Heap[self.root] # assin vaible to max node which is root node
        self.Heap[self.root] = self.Heap[self.cur_size]
        self.cur_size -= 1 # remove the location of node from size
        self.max_heapify(self.root) # travese with heapify() to maintian a heap structure
        return last 
        
    # THE Extract_Heap_Max FUNCTION to Remove any Node
    def Extract_Heap_Max(self,Node):
        anyNode = self.Heap[Node]  #assign varible to given node which is from user input in form of parameter of function
        self.Heap[self.root] = self.Heap[self.cur_size]
        self.cur_size -= 1 # remove the location of node from size
        self.max_heapify(self.root) # travese with heapify() to maintian a heap structure
        return anyNode
    
    # THE Heap_Increase_Key FUNCTION to increase to value stored at any node.
    def Heap_Increase_Key(self):
        H = self.Heap # input heap
        # Create the heap
        heapq.heapify(H) # used heapq library to implement a replacement fucntion 
        # Replace an element
        heapq.heapreplace(H,6) # increase value from 3 to 6
        print(H)
 
    # THE BUILD_HEAP FUNCTION
    def build_heap(self): 
        for i in range(self.cur_size//2, 0, -1): 
            self.max_heapify(i) # traverse whole array with heaapify function to maintain heap structure.
 
    # helper function to print the heap
    def print_heap(self):
        for i in range(1, (self.cur_size//2)+1): # iteratively print whole heap 
            print(self.Heap)
 
 #Driver Code

maxHeap = max_heap(10)
maxHeap.Insert_MaxHeap(15)
maxHeap.Insert_MaxHeap(7)
maxHeap.Insert_MaxHeap(9)
maxHeap.Insert_MaxHeap(4)
maxHeap.Insert_MaxHeap(13)
maxHeap.print_heap()
maxHeap.Build_Heap_Max()
maxHeap.print_heap()
maxHeap.Extract_Heap_Max(3)
maxHeap.print_heap()
maxHeap.Heap_Increase_Key()
```
